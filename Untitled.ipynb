```python
## Rename Columns (Clean)

df.columns = [
    "Date", 
    "CBP_Apprehended",
    "CBP_In_Custody",
    "CBP_Transferred",
    "HHS_In_Care",
    "HHS_Discharged"
]
```

```python
df.columns
```

```python
## Cleaning Data 

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
# Fill missing values

df.fillna(method="ffill", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
cols = [
    "CBP_Apprehended",
    "CBP_In_Custody",
    "CBP_Transferred",
    "HHS_In_Care",
    "HHS_Discharged"
]

for col in cols:
    df[col] = (
        df[col]
        .astype(str)                     # convert everything to string first
        .str.replace(",", "")           # remove commas
        .str.strip()                    # remove spaces
    )
    
    df[col] = pd.to_numeric(df[col], errors="coerce")  # convert to numeric
```

```python
df[cols].isnull().sum()
```

```python
df[cols] = df[cols].fillna(method="ffill")
```

### Feature Engineering (KPIs)

```python
df.columns
```

```python
#Transfer Efficiency

df['Transferred_Efficiency'] = df['CBP_Transferred'] / df['CBP_In_Custody']
```

```python
#Discharge_Efficiency

df["Discharge_Efficiency"] = df["HHS_Discharged"] / df["HHS_In_Care"]
```

```python
# Throughput

df['Throughput'] = df['HHS_Discharged'] / df['CBP_Apprehended']
```

```python
# Backlog

df['Backlog'] = (df ['CBP_In_Custody'] + df['HHS_In_Care']) - df['HHS_Discharged']
```

```python
df.head()
```

## visualization

```python
df.columns
```

```python
# Inflow vs Outflow

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['CBP_Apprehended'], label='Apprehended')
plt.plot(df['Date'], df['HHS_Discharged'], label='Discharged')
plt.title("Inflow vs Outflow")
plt.legend()
plt.show()
```

- Insight: If inflow > outflow → backlog grows

```python
df.columns
```

```python
# Efficiency Trends

df['Transfer_7d'] = df['Transferred_Efficiency'].rolling(7).mean()
df['Discharge_7d'] = df['Discharge_Efficiency'].rolling(7).mean()

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Transfer_7d'], label='Transferred Efficiency')
plt.plot(df['Date'], df['Discharge_7d'], label='Discharge Efficiency')
plt.legend()
plt.title("Efficieny Trends (7-Day Avg) ")
plt.show()
```

```python
##Backlog Analysis 

plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Backlog'])
plt.title('Backlog Over Time')
plt.show()
```

```python
# Bottleneck Detection  

threshold = df['Backlog'].quantile(0.9)
bottleneck = df[df["Backlog"] > threshold]

bottleneck[['Date', 'Backlog']]
```

```python
# Correlation Heatmap

plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
